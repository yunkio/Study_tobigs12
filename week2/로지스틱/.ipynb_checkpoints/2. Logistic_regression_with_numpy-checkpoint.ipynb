## 2. Sigmoid function
- Sigmoid function, Logistic function 

```python
def sigmoid(z):
    return 1 / (1 + np.exp(-z))
```

## 3. Hypothesis function 
- Sigmoid Function에는 z값이 들어가는 반면
- Hypothesis function에서는 theta와 x의 Linear Combination값이 들어간다

```python
def hypothesis_function(x, theta):
    z = (np.dot(x,theta)) # dot product를 해준다
    return sigmoid(z)
```

## 4. Cost function 
- Metrics의 Shape에 주의한다

### Loss function, Cost function
- 우리는 Loss function J를 구할 것이다

## $$ \mbox log 변환  $$

## $$ L^*=log(L) = ylog\sigma(x,B) + (1-y)log(1-\sigma(x,B)) $$ 

## $$ \mbox Convex\  function으로\  만들기\  위한 \ (-) 곱   $$

## $$ \mbox J는\ Loss\ Function  $$

## $$ J= - L^* $$

## $$ \mbox Loss\ function \ = \ Negative\ log \ likelihood\  $$

## $$ \mbox N개의 \ 데이터에\ 대한\ Negative\ log \ Likelihood  $$
## $$ J(x_i,y_i|B) = \frac{\mathrm{1} }{\mathrm{N}}(-y_i\sum_{i=1}^Nlog\sigma(x_i,B)-(1-y_i)\sum_{i=1}^Nlog(1-\sigma(x_i,B))) $$ 


## 위의 수식을 코드로 구현하면 된다!

```python
# y_data.shape[0]
# y_data.shape[0]은 200이다.
```

```python
# metrics의 shape에 주목하자!

def compute_cost(x, y, theta):  # J는 cost function, theta는 위 수식의 B와 같다.
    m = y.shape[0]
    J = (-1.0 / m )* (
        y.T.dot(np.log(hypothesis_function(x,theta))) + \
        (1-y).T.dot(np.log(1- hypothesis_function(x,theta))))
    
    return (-1.0 /m )* (y * np.log(hypothesis_function(x,theta)) + (1-y) \
            * np.log(1- hypothesis_function(x,theta))).sum()
```

## 5. Weight update
 - partial_margin
 - delta

```python
def minimize_gradient(x, y, theta, iterations=100000, alpha=0.01):
    m = y.size
    cost_history = []
    theta_history = []
    
    
    for _ in range(iterations):
        original_theta = theta
        for i in range(theta.size):
            partial_marginal = x[:, i].reshape(x.shape[0], 1)
            delta = hypothesis_function(x, original_theta) - y
            grad_i = delta.T.dot(partial_marginal)
#             print(partial_marginal)
            theta[i] = theta[i] - (alpha  * grad_i)
        
        if (_ % 100) == 0:
            theta_history.append(theta)
            cost_history.append(compute_cost(x, y, theta))

    return theta, np.array(cost_history), np.array(theta_history)
```

### 이걸 다 알아야 할까?!
Tensor flow에서는 GradientDescentOptimize()에서 한번에 구해준다...

Theta의 사이즈 만큼 Iteration하면서 Theta값이 업데이트 된다는 것만 알아두자!

```python
theta_initial = np.ones([3,1])
theta, cost_history, theta_history = minimize_gradient(
    x_data, y_data,theta_initial, 10000, 0.001)
```

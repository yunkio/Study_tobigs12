# 1. EDA 및 전처리

```python
df = pd.read_csv("Mall_Customers.csv")
```

# 2. 클러스터링 적용해보기

## 2-1. K-means

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
# 최적의 k

X = df
distorsions = []
for k in range(2, 20):
    kmeans = KMeans(n_clusters=k)
    kmeans.fit(X)
    distorsions.append(kmeans.inertia_)

fig = plt.figure(figsize=(15, 5))
plt.plot(range(2, 20), distorsions)
plt.grid(True)
plt.title('Elbow curve')
```

```python
#K-Means 군집 분석 
model = KMeans(n_clusters=6,algorithm='auto')
model.fit(df)
predict = pd.DataFrame(model.predict(df))
predict.columns=['predict']

# predict 추가 
r = pd.concat([df,predict],axis=1)
r
```

```python
# 각 군집의 중심점 
centers = pd.DataFrame(model.cluster_centers_,columns=['Gender','Age','Annual Income (k$)','Spending Score (1-100)'])
center_x = centers['Age']
center_y = centers['Annual Income (k$)']
center_z = centers['Spending Score (1-100)']

from mpl_toolkits.mplot3d import Axes3D
# scatter plot
fig = plt.figure( figsize=(6,6))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(r['Spending Score (1-100)'],r['Age'],r['Annual Income (k$)'],c=r['predict'],alpha=0.5)
ax.scatter(center_x,center_y,center_z,s=50,marker='D',c='r')
ax.set_xlabel('Spending Score')
ax.set_ylabel('Age')
ax.set_zlabel('Annual Income')
plt.show()
```

```python
r["predict"].value_counts()

#2    45
#0    39
#3    38
#1    35
#4    22
#5    21
```

## 2-2. DBSCAN

```python
from sklearn.cluster import DBSCAN
import matplotlib.pyplot  as plt
import seaborn as sns
```

```python
model = DBSCAN(eps=11,min_samples=5)

predict = pd.DataFrame(model.fit_predict(df))
predict.columns=['predict']

# concatenate labels to df as a new column
r = pd.concat([df,predict],axis=1)
r["predict"].value_counts()

# 2    80
#-1    55
# 3    32
# 0    18
# 4    10
# 1     5
```

```python
from mpl_toolkits.mplot3d import Axes3D
# scatter plot
fig = plt.figure( figsize=(6,6))
ax = Axes3D(fig, rect=[0, 0, .95, 1], elev=48, azim=134)
ax.scatter(r['Spending Score (1-100)'],r['Age'],r['Annual Income (k$)'],c=r['predict'],alpha=0.5)
ax.scatter(center_x,center_y,center_z,s=50,marker='D',c='r')
ax.set_xlabel('Spending Score')
ax.set_ylabel('Age')
ax.set_zlabel('Annual Income')
plt.show()
```

```python
# 원래 label이 되어있는 데이터가 아니기 때문에 평가에 어려움이 있다.
# 단 DBSCAN의 경우 파라미터(eps, min_samples)를 잘 조정해보아도 k-means에 비해 시각적으로 덜 분류되는 모습을 볼 수 있었다.
```

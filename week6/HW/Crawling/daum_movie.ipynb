```python
import time
from time import sleep
import re
from tqdm import tqdm
from random import randint
import json
import urllib.request
import csv
import pandas as pd

# Import BeautifulSoup and requests module
from bs4 import BeautifulSoup as bs
import requests
```

```python
def totalaudience(movieid):
        url = "http://movie.daum.net/moviedb/main/totalAudience.json?movieId=" + str(movieid)
        response = urllib.request.urlopen(url)
        json_data = response.read()
        data = json.loads(json_data.decode('utf-8'))
        return data['totalAudience']
    
#참고 : https://kin.naver.com/qna/detail.nhn?d1id=1&dirId=10402&docId=300841549&qb=L21vdmllZGIvbWFpbi90b3RhbEF1ZGllbmNlLmpzb24=&enc=utf8§ion=kin&rank=1&search_sort=0&spq=0&pid=UTd9cwp0YihssAlKkz4ssssst70-319291&sid=oCkwxx%2BUaBkh6QJig%2F8ZbQ%3D%3D
```

```python
def daum_movie_crawler(year):
    headers = {'User-Agent': 'Mozilla/5.0 (Macintosh; InteSl Mac O X 10_10_1) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/39.0.2171.95 Safari/537.36'}
    main_url = 'https://movie.daum.net/boxoffice/yearly?year=' + str(year)
    
    # Initialize 
    data = []
    cnt= 0 
    
    response = requests.get(main_url,headers=headers)
    body_table = bs(response.text,'lxml')
    poster_table = body_table.find_all('div',class_='wrap_movie')
    
    #포스터 테이블을 looping 하며 제목 찾기
    # Trye-except 문을 이용하기 
    for poster in poster_table:
        try:
            title = poster.find('a', class_='name_movie #title').text
            grade_table = poster.find("span", {"class":"wrap_grade grade_netizen"}) 
            grade = float("".join(re.findall('m_\d(\d)', str(grade_table))))/10 # 관람객평점
            movieid = re.findall('=(\d+)', poster.find('a')['href'])
            audience = totalaudience(movieid[0]) # 누적 관람객
            critic_table = poster.find("span", {"class":"wrap_grade grade_critic"})
            critic = float("".join(re.findall('m_\d(\d)', str(critic_table))))/10 # 전문가평점
            data.append([title, grade, critic, audience])                                 
                                                      
        except Exception as e :
            data.append([title, grade, 'NA', audience])
            print(e,'Error at {} for {}'.format(cnt,title))

    time.sleep(1)
    return data
```

```python
total_data = list()
for year in [2018,2019]:
    total_data.extend(daum_movie_crawler(year))
```
